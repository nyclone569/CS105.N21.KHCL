```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
from google.colab.patches import cv2_imshow
```

```python
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascades/haarcascade_frontalcatface.xml')
eye_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascades/haarcascade_eye.xml')
mouth_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascades/haarcascade_smile.xml')

img = cv2.imread('/content/drive/MyDrive/F1.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.1,4) 

for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
  eyes = eye_cascade.detectMultiScale(gray, 1.1,4) 
  for (xx,yy,ww,hh) in eyes:
    cv2.rectangle(img,(xx,yy),(xx+ww,yy+hh),(0,255,0),2)
  mouths= mouth_cascade.detectMultiScale(gray, 1.1,4) 
  for (xxx,yyy,www,hhh) in mouths:
    cv2.rectangle(img,(xxx,yyy),(xxx+www,yyy+hh),(0,0,255),2)

cv2_imshow(img)
cv2.waitKey()
```
